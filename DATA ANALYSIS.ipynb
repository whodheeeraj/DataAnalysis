```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# loading dataset
epl_data = pd.read_csv('EPL_20_21.csv')
epl_data.head()
```

```python
epl_data.info()
```

```python
epl_data.describe()
```

```python
epl_data.isna().sum()
```

```python
# creating coloumn for how many minutes a player played a match
epl_data['MinsPerMatch'] = (epl_data['Mins']/epl_data['Matches']).astype(int)

# creating coloumn for how many goal scored in a match
epl_data['GoalsPerMatch'] = (epl_data['Goals']/epl_data['Matches']).astype(float)
```

```python
epl_data.head()
```

```python
# total goals scored in entire season

TotalGoals = epl_data['Goals'].sum()
TotalGoals
```

```python
# total penalty goals scored
TotalPenGoals = epl_data['Penalty_Goals'].sum()
TotalPenGoals
```

```python
# total penalty attempted
TotalPenAttempts = epl_data['Penalty_Attempted'].sum()
TotalPenAttempts
```

```python
# goal scoring ratio of penalty
PenaltyRatio = (epl_data['Penalty_Attempted'].sum()/epl_data['Penalty_Goals'].sum()).astype(float)
PenaltyRatio
```

```python
#  pie chart for penalty missed vs scored
# import seaborn as sns

plt.figure(figsize=(13,6))
plt_not_scored = TotalPenAttempts - TotalPenGoals
df = [plt_not_scored, TotalPenGoals]
labels = ['Penalties Missed', 'Penalties Scored']
# color = sns.color_palette('set2')
plt.pie(df, labels = labels, autopct = '%.0f%%')
plt.show()
```

```python
#players based on age group

Under20 = epl_data[epl_data['Age'] <= 20]
age20_25 = epl_data[(epl_data['Age'] > 20) & (epl_data['Age'] <= 25)]
age25_30 = epl_data[(epl_data['Age'] > 25) & (epl_data['Age'] <= 30)]
above30 = epl_data[epl_data['Age'] > 30]
```

```python
plt.figure(figsize=(12,6))
x = np.array([Under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), above30['Name'].count()])
mylabels = ["Under 20", "B/w 20-25", "B/w 25-30", "Above 30"]
plt.title('Total Players with Age Group', fontsize = 20)
plt.pie(x, labels = mylabels, autopct = "%.1f%%")
plt.show
```

```python
#Total Assists from each club

AssistByClubs = pd.DataFrame(epl_data.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style = 'whitegrid', color_codes = True)
Ac = sns.barplot(x = 'Club', y = 'Assists', data = AssistByClubs.sort_values(by='Assists'), palette='Set2')
Ac.set_xlabel("Club", fontsize = 30)
Ac.set_ylabel("Assists", fontsize = 30)
plt.xticks(rotation = 60)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of clubs vs Total Assists', fontsize = 30)
```

```python
# Top ten assist providers
top_10_assists = epl_data[['Name', 'Club', 'Assists', 'Matches']].nlargest(n=10, columns = 'Assists')
top_10_assists
```

```python
GoalsByClubs = pd.DataFrame(epl_data.groupby('Club', as_index = False)['Goals'].sum())
sns.set_theme(style = 'whitegrid', color_codes = True)
Ac = sns.barplot(x = 'Club', y = 'Goals', data = GoalsByClubs.sort_values(by='Goals'), palette='rocket')
Ac.set_xlabel("Club", fontsize = 30)
Ac.set_ylabel("Goals", fontsize = 30)
plt.xticks(rotation = 60)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of clubs vs Total Goals', fontsize = 30)
```
